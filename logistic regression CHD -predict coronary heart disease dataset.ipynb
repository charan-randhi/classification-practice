# logistic regression model

```python
y=df['TenYearCHD']
x=df.drop('TenYearCHD',axis=1)
```

```python
import statsmodels.api as sm
xc=sm.add_constant(x)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
```

```python
pd.DataFrame([VIF(xc.values,i) for i in range(xc.shape[1])],index=xc.columns,columns=['vif'])
```

```python
model=sm.Logit(y,xc).fit()
model.summary()
```

```python
list(xc.columns)
```

```python
cols = ['const',
 'male',
 'age',
 'cigsPerDay',
 'prevalentStroke',
 'prevalentHyp',
 'diabetes',
 'totChol',
 'sysBP',
 'glucose']

model_1 = sm.Logit(y,xc.loc[:,cols]).fit()
model_1.summary()
```

```python
cols = ['const',
 'male',
 'age',
 'cigsPerDay',
 'prevalentStroke',
 'sysBP',
 'glucose']

model_1 = sm.Logit(y,xc.loc[:,cols]).fit()
model_1.summary()
```

```python
params=pd.DataFrame(model_1.params,columns=['coeft'])
params['exp(coeft)']=np.exp(model_1.params)
params
```

```python
xc2=xc.loc[:,cols]
xc2['chd']=y

xc2.to_csv('chd1.csv')
```

```python
prob= model_1.predict(xc.loc[:,cols])
prob=pd.DataFrame(prob,columns=['prob'])
prob['chd']=y
prob['y_est']=prob['prob'].apply(lambda x:0 if x<0.5 else 1)
prob
```

```python
prob['y_est'].value_counts()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(prob['chd'],prob['y_est'])
```

```python
accuracy_score(prob['chd'],prob['y_est'])
```

```python
plt.figure(figsize=(12,8))
sns.distplot(prob['prob'][prob['chd']==0],color='r')
sns.distplot(prob['prob'][prob['chd']==1],color='g')
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(prob['chd'],prob['y_est']))
```

# ML model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='liblinear')
```

```python
lr.fit(x_train,y_train)
```

```python
y_train_pred=lr.predict(x_train)
y_train_pred
```

```python
y_train_prob=lr.predict_proba(x_train)[:,1]
```

```python
y_train_prob
```

```python
y_train_prob[:,1] #to get the predicted prob of 1
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("accuracy of train : ",accuracy_score(y_train,y_train_pred))
```

```python
print("confusion matrix train : ",'\n', confusion_matrix(y_train,y_train_pred))
```

```python
y_test_pred=lr.predict(x_test)
y_test_pred
```

```python
y_test_prob= lr.predict_proba(x_test)[:,1]
y_test_prob
```

```python
print("accuracy of train : ",accuracy_score(y_test,y_test_pred))
```

```python
print("confusion matrix test : ",'\n', confusion_matrix(y_test,y_test_pred))
```

```python
cm=confusion_matrix(y_train,y_train_pred)
tpr=cm[1,1]/((cm[1,0])+(cm[1,1]))
fpr=cm[0,1]/((cm[0,0])+(cm[0,1]))
```

```python
print("TPR : ",tpr, "fpr : ",fpr)
```

```python
prb_train = pd.DataFrame(y_train_prob,columns=["prob"])
prb_train["y_pred"] = prb_train["prob"].apply(lambda x:0 if x<0.5 else 1)
print(prb_train.head())
cm_train = confusion_matrix(y_train,prb_train["y_pred"])
TPR = cm_train[1,1]/(cm_train[1,0] + cm_train[1,1])
FPR = cm_train[0,1]/(cm_train[0,0] + cm_train[0,1])
print("TPR for Train data:",TPR)
print("FPR for Train data:",FPR)
```

```python
prb_train = pd.DataFrame(y_train_prob,columns=["prob"])
prb_train["y_pred"] = prb_train["prob"].apply(lambda x:0 if x<0.4 else 1)
print(prb_train.head())
cm_train = confusion_matrix(y_train,prb_train["y_pred"])
TPR = cm_train[1,1]/(cm_train[1,0] + cm_train[1,1])
FPR = cm_train[0,1]/(cm_train[0,0] + cm_train[0,1])
print("TPR for Train data:",TPR)
print("FPR for Train data:",FPR)
```

```python
prb_train = pd.DataFrame(y_train_prob,columns=["prob"])
prb_train["y_pred"] = prb_train["prob"].apply(lambda x:0 if x<0.1 else 1)
print(prb_train.head())
cm_train = confusion_matrix(y_train,prb_train["y_pred"])
TPR = cm_train[1,1]/(cm_train[1,0] + cm_train[1,1])
FPR = cm_train[0,1]/(cm_train[0,0] + cm_train[0,1])
print("TPR for Train data:",TPR)
print("FPR for Train data:",FPR)
```
